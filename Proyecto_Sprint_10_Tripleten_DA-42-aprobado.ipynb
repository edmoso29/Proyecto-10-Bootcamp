#### 2.2.9 - Encuentra la significancia estad√≠stica de la diferencia en la conversi√≥n entre los grupos utilizando los datos en bruto. Haz conclusiones y conjeturas.

```python
conversi√≥n_por_grupo = orders_clean.groupby('group')['visitor_id'].nunique()
visitas_por_grupo = visits.groupby('group')['visits'].sum()
```

```python
# Extraer los datos para successes y trials
successes = [conversi√≥n_por_grupo['A'], conversi√≥n_por_grupo['B']]
trials = [visitas_por_grupo['A'], visitas_por_grupo['B']]

# Calcular proporci√≥n de √©xito de cada grupo
p1 = successes[0] / trials[0]
p2 = successes[1] / trials[1]
difference = p2 - p1

# Calcular la proporci√≥n de √©xito combinada
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

# Calcula la estad√≠stica en desviaciones est√°ndar de la distribuci√≥n normal est√°ndar
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

# Establecer la distribuci√≥n normal est√°ndar (media 0, desviaci√≥n est√°ndar 1)
distr = st.norm(0, 1) 

# Calcular el valor p (prueba de dos colas)
p_value = (1 - distr.cdf(abs(z_value))) * 2

print(f"z_value: {z_value:.4f}")
print(f"p_value: {p_value:.4f}")

alpha = 0.05

if p_value < alpha:
    print("Rechazar la hip√≥tesis nula: hay una diferencia significativa entre las proporciones.")
else:
    print("No se pudo rechazar la hip√≥tesis nula: no hay raz√≥n para pensar que las proporciones son diferentes.")
```
